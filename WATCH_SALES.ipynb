### Now let us divide the date into separate date and day value to analyse the price w.r.t year

```python
### Separating the day and year value from date column in file1

YEAR=[]
DAY=[]
for i in file1['ENTRY_DATE']:
    x=i.split('-')
    YEAR.append(x[0])
    DAY.append(x[1])

file1['YEAR']=YEAR
file1['DAY']=DAY
file1=file1.drop(['ENTRY_DATE'],axis=1)

for x in range(file1.shape[0]):
    if file1.loc[x,'YEAR']=='':
        file1.loc[x,'YEAR']=int(file1.loc[x-1,'YEAR'])
    else:
        file1.loc[x,'YEAR']=int(file1.loc[x,'YEAR'])
                
```

Find feature importance of numerical column

```python
dt=input_data2.loc[:,['TRANS POSTED WK','TRANS ACTUAL MONTH','TRANS ACTUAL DAY','TRANS ACTUAL WK', 'NET SALES UNITS','UNIT PRICE']]
dt.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing

lab_enc = preprocessing.LabelEncoder()
output_encoded = lab_enc.fit_transform(output_file2)
# Build the model
rfc = RandomForestClassifier()

# Fit the model
rfc.fit(dt,output_encoded)
names=dt.columns
for feature in zip(names, rfc.feature_importances_):
    print(feature)
```

# DEALING WITH CATEGORICAL DATA

```python
### Converting categorical data to numerical one using one_hot_encoding

one_hot_brand = pd.get_dummies(input_data2['BRAND'])
input_data2=input_data2.drop(['BRAND'], axis = 1)
```

```python
### Converting categorical data to numerical one using one_hot_encoding

one_hot_loc_code = pd.get_dummies(input_data2['LOCATION CODE'])
input_data2=input_data2.drop(['LOCATION CODE'], axis = 1)
```

```python
### Converting categorical data to numerical one using one_hot_encoding

one_hot_sales_activity_code = pd.get_dummies(input_data2['SALES ACTIVITY CODE'])
input_data2=input_data2.drop(['SALES ACTIVITY CODE'], axis = 1)
```

```python
### Concating all of them together

frames = [input_data2, one_hot_brand,one_hot_loc_code,one_hot_sales_activity_code]
input_ = pd.concat(frames,axis=1,sort=False)
input_data=input_.drop(['INVOICED AMOUNT '], axis = 1)
input_.head()
```
